```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split as tts
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
```

```python
# Splitting the data into training , testing and validation sets
train_val_df,test_df = tts(df,test_size = 0.20, random_state = 42)
train_df, val_df = tts(train_val_df,test_size = 0.20,random_state = 42)
```

```python
train_df
```

```python
inputs = df.columns[:-1]
inputs
```

```python
target = ['Price_for_one']
target
```

```python
train_inputs = train_df[inputs]
train_outputs = train_df[target]
val_inputs = val_df[inputs]
val_outputs = val_df[target]
test_inputs = test_df[inputs]
test_outputs = test_df[target]
```

```python
numeric = train_inputs.select_dtypes(np.number).columns.tolist()
categorical = train_inputs.select_dtypes('object').columns.tolist()
```

### Preprocessing

```python
#Scaling

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(df[numeric])
```

```python
train_inputs[numeric] = scaler.transform(train_inputs[numeric])
val_inputs[numeric] = scaler.transform(val_inputs[numeric])
test_inputs[numeric] = scaler.transform(test_inputs[numeric])
```

```python
#Encoding
from sklearn.preprocessing import OneHotEncoder
```

```python
# Assuming 'categorical' is a list of your categorical column names
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore').fit(df[categorical])
```

```python
encoded_cols = list(encoder.get_feature_names_out(categorical))
```

```python
encoded_cols 
```

```python
train_inputs[encoded_cols] = encoder.transform(train_inputs[categorical])
val_inputs[encoded_cols] = encoder.transform(val_inputs[categorical])
test_inputs[encoded_cols] = encoder.transform(test_inputs[categorical])
```

```python
train_inputs = train_inputs[numeric + encoded_cols]
val_inputs = val_inputs[numeric + encoded_cols]
test_inputs = test_inputs[numeric + encoded_cols]
```

### Fitting and Evaluation

```python
def get_user_input():
    
    input_cuisine = input("Enter your preferred cuisine: ")

    input_location = input("Enter your preferred location: ")
    try:
        input_preferred_price = float(input("Enter your preferred price for 1: "))
    except ValueError:
        print("Please enter a valid number for price.")
        return None, None, None  # Return None if invalid input for price

    return input_cuisine, input_location, input_preferred_price

# Call the function to get user inputs
input_cuisine, input_location, input_preferred_price = get_user_input()

if input_cuisine and input_location and input_preferred_price is not None:
    print(f"Inputs - Cuisine: {input_cuisine}, Location: {input_location}, Preferred Price: {input_preferred_price}")
else:
    print("Invalid inputs received.")
```
